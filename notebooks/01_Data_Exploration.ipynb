```python
import os
```

```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
```

```python
import random
```

```python
train_path = "../data/Training"
```

```python
print(train_path)
```

```python
classes = os.listdir(train_path)

print(classes)
```

```python
for cls in classes:

    folder = os.path.join(train_path, cls)

    number = len(os.listdir(folder))

    print(cls, number)
```

```python
glioma_folder = os.path.join(train_path, "glioma")

print(glioma_folder)
```

```python
glioma_images = os.listdir(glioma_folder)

print(len(glioma_images))
```

```python
random_image = random.choice(glioma_images)

print(random_image)
```

```python
image_path = os.path.join(glioma_folder, random_image)

print(image_path)
```

```python
image = Image.open(image_path)
```

```python
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```
