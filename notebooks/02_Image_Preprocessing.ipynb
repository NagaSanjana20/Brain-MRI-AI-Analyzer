```python
import numpy as np

import matplotlib.pyplot as plt

from PIL import Image

import random

import os
```

```python
train_path = "../data/Training"

glioma_folder = os.path.join(train_path, "glioma")

glioma_images = os.listdir(glioma_folder)

random_image = random.choice(glioma_images)

image_path = os.path.join(glioma_folder, random_image)

image = Image.open(image_path)
```

```python
plt.imshow(image, cmap="gray")
plt.title("Original MRI")
plt.axis("off")
plt.show()
```

```python
resized_image = image.resize((224,224))
```

```python
print(image.size)

print(resized_image.size)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(resized_image, cmap="gray")
plt.title("Resized")
plt.axis("off")

plt.show()
```

```python
rotated = image.rotate(90)
plt.imshow(rotated, cmap="gray")
plt.title("Rotated 90°")
plt.axis("off")
plt.show()
```

```python
flipped = image.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(flipped, cmap="gray")
plt.title("Flipped")
plt.axis("off")
plt.show()
```

```python
cropped = image.crop((100,100,400,400))
plt.imshow(cropped, cmap="gray")
plt.title("Cropped")
plt.axis("off")
plt.show()
```
